```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import spacy

df=pd.read_csv('processed_text.csv')

# Load the English model of spaCy
nlp = spacy.load('en_core_web_sm')

# Concatenate all the texts into a single long string
text = ' '.join(df['text'])

# Create a Doc object
doc = nlp(text)
```

```python
# Define a list of words to exclude
exclude_words = ['ewallet']

# Keep only adjectives and exclude specific words
words = [token.text for token in doc if token.pos_ in ['ADJ'] and token.text not in exclude_words]

# Count word frequency
word_counts = Counter(words)

# Find the top 10 most common words
common_words = word_counts.most_common(10)

# Set the size of the figure
plt.figure(figsize=(10, 5))

# Plot the bar chart
plt.bar(*zip(*common_words))

# Rotate the x-axis labels
plt.xticks(rotation=45)

# Set the x-axis and y-axis labels
plt.xlabel('Words')
plt.ylabel('Counts')

# Display the plot
plt.show()
```

```python
# Keep only nouns and exclude specific words
words = [token.text for token in doc if token.pos_ in ['NOUN']]

# Count word frequency
word_counts = Counter(words)

# Find the top 10 most common words
common_words = word_counts.most_common(10)

# Set the size of the figure
plt.figure(figsize=(10, 5))

# Plot the bar chart
plt.bar(*zip(*common_words))

# Rotate the x-axis labels
plt.xticks(rotation=45)

# Set the x-axis and y-axis labels
plt.xlabel('Words')
plt.ylabel('Counts')

# Display the plot
plt.show()
```
