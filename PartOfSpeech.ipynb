```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import spacy

df=pd.read_csv('processed_text.csv')
```

```python
# 加载spaCy的英语模型
nlp = spacy.load('en_core_web_sm')

# 将所有的文本连接成一个长字符串
text = ' '.join(df['text'])

# 创建一个Doc对象
doc = nlp(text)

# 定义一个列表，包含你想排除的词
exclude_words = ['ewallet']

# 只保留形容词，并排除特定的词
words = [token.text for token in doc if token.pos_ in ['ADJ'] and token.text not in exclude_words]

# 统计词频
word_counts = Counter(words)

# 找出出现次数最多的10个词
common_words = word_counts.most_common(10)

# 设置图形的大小
plt.figure(figsize=(10, 5))

# 绘制条形图
plt.bar(*zip(*common_words))

# 旋转x轴的标签
plt.xticks(rotation=45)

# 设置x轴和y轴的标签
plt.xlabel('Words')
plt.ylabel('Counts')

# 显示图形
plt.show()
```
